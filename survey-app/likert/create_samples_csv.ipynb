```python
import os
import pandas as pd

# Path to your samples folder
root_dir = "samples"

# Prepare list to collect sample data
samples_data = []

# Walk through each model subfolder
for model_folder in sorted(os.listdir(root_dir)):
    model_path = os.path.join(root_dir, model_folder)
    
    if os.path.isdir(model_path):
        # Get all mp3 files
        files = sorted(os.listdir(model_path))
        
        # Loop through each file
        for file in files:
            if file.endswith(".mp3"):
                # Extract index number from filename
                index = file.split("_")[0]
                
                # Prepare sample_id
                sample_id = f"{model_folder}_{int(index):04d}"
                
                # Prepare model_name
                model_name = model_folder
                
                # Prepare relative file path
                file_path = os.path.join("samples", model_folder, file).replace("\\", "/")
                
                samples_data.append({
                    "sample_id": sample_id,
                    "model_name": model_name,
                    "file_path": file_path
                })

# Convert to DataFrame
samples_df = pd.DataFrame(samples_data)

# Sort for cleanliness
samples_df = samples_df.sort_values(["model_name", "sample_id"])

# Preview
samples_df.head()
```

```python
import pandas as pd
import os

# Assume samples_df already exists from your previous code!

# Path to your master CSV
master_csv_path = "samples.csv"

if os.path.exists(master_csv_path):
    existing_samples_df = pd.read_csv(master_csv_path, dtype={"sample_id": str})
    
    # Merge the new data
    combined_df = pd.concat([existing_samples_df, samples_df], ignore_index=True)
    
    # Drop duplicate sample_ids
    combined_df = combined_df.drop_duplicates(subset=["sample_id"])
else:
    combined_df = samples_df

# Save merged result
combined_df.to_csv(master_csv_path, index=False)

print("✅ samples.csv has been updated successfully!")
```
