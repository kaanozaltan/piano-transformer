```python
filepath = '../data/maestro/2004/MIDI-Unprocessed_SMF_02_R1_2004_01-05_ORIG_MID--AUDIO_02_R1_2004_05_Track05_wav.midi'
```

To get 

```python
from mido import MidiFile

midi = MidiFile(filepath)

for i, track in enumerate(midi.tracks):
    print(f'Track {i}: {track.name}')
    for msg in track:
        print(msg)
```

```python
print(f"MIDI type: {midi.type}")
print(f"Number of tracks: {len(midi.tracks)}")
print(f"Length of MIDI file: {midi.length/60} minutes")
print(f"Ticks per beat: {midi.ticks_per_beat}")
print(f"Number of notes: {sum(1 for track in midi.tracks for msg in track if msg.type == 'note_on')}")
print(f"Number of control changes: {sum(1 for track in midi.tracks for msg in track if msg.type == 'control_change')}")
print(f"Number of program changes: {sum(1 for track in midi.tracks for msg in track if msg.type == 'program_change')}")
print(f"Number of pitch bends: {sum(1 for track in midi.tracks for msg in track if msg.type == 'pitchwheel')}")
```

```python
import pretty_midi

midi = pretty_midi.PrettyMIDI(filepath)

n=10

for i, instrument in enumerate(midi.instruments):
    print(f"Instrument {i}: {instrument.name}")
    for note in instrument.notes[:n]:
        print(
            f"Note {note.pitch}, start: {note.start:.2f}s, end: {note.end:.2f}s, velocity: {note.velocity}"
        )
```
